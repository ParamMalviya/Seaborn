```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
var_1 = [1,2,3,4,5,6,7,8]
var_2 = [2,3,4,5,1,6,7,8]
```

```python
plt.plot(var_1,var_2)
plt.show()
```

```python
import pandas as pd
```

```python
df = pd.DataFrame({'var_1':var_1, 'var_2':var_2})
sns.lineplot(x='var_1', y='var_2', data= df)
plt.show()
```

```python
y_1 = sns.load_dataset("penguins")
y_1
```

```python
sns.lineplot(x="bill_length_mm", y="bill_depth_mm", data=y_1, hue="sex", size=10, style='sex', palette="Accent", markers=["o",">"])
plt.show()
```

```python
y_2=y_1.head(20)
y_2
```

```python
sns.lineplot(x="bill_length_mm", y="bill_depth_mm", data=y_2, hue="sex", size=10, style='sex', palette="Accent", markers=["o",">"], dashes=True)
plt.show()
```

```python
sns.lineplot(x="bill_length_mm", y="bill_depth_mm", data=y_2, hue="sex", size=10, style='sex', palette="Accent", markers=["o",">"], dashes=False, 
legend=False)
plt.title("Lineplot")
plt.grid()
plt.show()
```
